## <font color='1957d1'> Trayectorias (ℝ → ℝⁿ) </font>

Una **trayectoria** (o curva paramétrica) es una función $\alpha: I \subseteq \mathbb{R} \to \mathbb{R}^n$ que asigna un vector en $\mathbb{R}^n$ a cada valor del parámetro $t \in I$. Geométricamente, traza una curva en el espacio.

Para definirla en Python:
1. Declarar la variable paramétrica con `sp.Symbol('t')`.
2. Definir cada componente como función de $t$ en un `sp.Matrix`.
3. Convertir a función numérica con `sp.lambdify` para graficar.

**Ejemplo 1.** Definamos y grafiquemos la elipse $\alpha(t) = (\cos t,\, 2\sin t)$, $t \in [0, 2\pi]$.

```python
# Variable paramétrica
t = sp.Symbol('t')

# Definición simbólica de la trayectoria
# alpha: [0, 2*pi] -> R²,  alpha(t) = (cos(t), 2*sin(t))
alpha = sp.Matrix([sp.cos(t), 2*sp.sin(t)])
print('α(t) =', alpha.T)

# Lambdificar componentes para graficar
alpha_num = sp.lambdify(t, list(alpha), 'numpy')

# Evaluar en el intervalo [0, 2π]
t_vals = np.linspace(0, 2*np.pi, 300)
x_pts, y_pts = alpha_num(t_vals)

# Graficar
fig, ax = plt.subplots(figsize=(5, 5))
ax.plot(x_pts, y_pts, color='#1957d1', linewidth=2)
ax.set_aspect('equal')
ax.set_title(r'$\alpha(t) = (\cos t,\; 2\sin t)$', fontsize=13)
ax.set_xlabel(r'$x$')
ax.set_ylabel(r'$y$')
ax.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

**Ejemplo 2.** En $\mathbb{R}^3$, definamos la hélice $\alpha(t) = (\cos t,\, \sin t,\, t)$, $t \in [0, 4\pi]$.

```python
# Trayectoria en R³: hélice
# alpha: [0, 4*pi] -> R³,  alpha(t) = (cos(t), sin(t), t)
alpha3 = sp.Matrix([sp.cos(t), sp.sin(t), t])
print('α(t) =', alpha3.T)

# Lambdificar
alpha3_num = sp.lambdify(t, list(alpha3), 'numpy')

# Evaluar en [0, 4π]
t_vals = np.linspace(0, 4*np.pi, 500)
x_pts, y_pts, z_pts = alpha3_num(t_vals)

# Graficar en 3D
fig = plt.figure(figsize=(6, 6))
ax = fig.add_subplot(111, projection='3d')
ax.plot(x_pts, y_pts, z_pts, color='#1957d1', linewidth=2)
ax.set_title(r'$\alpha(t) = (\cos t,\; \sin t,\; t)$', fontsize=13)
ax.set_xlabel(r'$x$')
ax.set_ylabel(r'$y$')
ax.set_zlabel(r'$z$')
plt.tight_layout()
plt.show()
```

## <font color='1957d1'> Campos Escalares (ℝⁿ → ℝ) </font>

Un **campo escalar** es una función $f: D \subseteq \mathbb{R}^n \to \mathbb{R}$ que asigna un número real a cada punto del dominio. Son los ejemplos más clásicos de funciones de varias variables.

Para definirlos y visualizarlos en Python:
1. Declarar las variables con `sp.symbols`.
2. Escribir la expresión simbólica.
3. Lambdificar y crear una malla con `np.meshgrid`.
4. Graficar la superficie 3D con `plot_surface` y las curvas de nivel con `contourf`.

**Ejemplo 3.** Definamos el paraboloide $f(x, y) = x^2 + y^2$.

```python
# Variables simbólicas
x, y = sp.symbols('x y')

# Campo escalar: paraboloide
# f: R² -> R,  f(x, y) = x² + y²
f = x**2 + y**2
print('f(x, y) =', f)

# Lambdificar para evaluación numérica
f_num = sp.lambdify((x, y), f, 'numpy')

# Crear malla de puntos
x_vals = np.linspace(-3, 3, 150)
y_vals = np.linspace(-3, 3, 150)
X, Y = np.meshgrid(x_vals, y_vals)
Z = f_num(X, Y)

# Gráficas
fig = plt.figure(figsize=(12, 5))

# Superficie 3D
ax1 = fig.add_subplot(121, projection='3d')
ax1.plot_surface(X, Y, Z, cmap='Blues', alpha=0.85)
ax1.set_title(r'$f(x,y) = x^2 + y^2$', fontsize=13)
ax1.set_xlabel(r'$x$'); ax1.set_ylabel(r'$y$'); ax1.set_zlabel(r'$z$')

# Curvas de nivel
ax2 = fig.add_subplot(122)
cp = ax2.contourf(X, Y, Z, levels=15, cmap='Blues')
plt.colorbar(cp, ax=ax2, label=r'$f(x,y)$')
ax2.contour(X, Y, Z, levels=15, colors='#1957d1', alpha=0.5, linewidths=0.8)
ax2.set_title(r'Curvas de nivel: $f(x,y) = c$', fontsize=13)
ax2.set_xlabel(r'$x$'); ax2.set_ylabel(r'$y$')
ax2.set_aspect('equal')

plt.tight_layout()
plt.show()
```

## <font color='1957d1'> Campos Vectoriales (ℝⁿ → ℝᵐ) </font>

Un **campo vectorial** es una función $\mathbf{F}: D \subseteq \mathbb{R}^n \to \mathbb{R}^m$ que asigna un vector en $\mathbb{R}^m$ a cada punto del dominio. Son fundamentales en física para representar campos de velocidad, fuerza, flujo, etc.

Para visualizarlos en Python (caso n = m = 2):
1. Crear una malla de puntos con `np.meshgrid`.
2. Calcular las componentes $U$ y $V$ del campo en cada punto.
3. Graficar con `plt.quiver(X, Y, U, V)`.
4. Para visualizar solo la dirección, normalizar dividiendo por la magnitud.

**Ejemplo 4.** El campo de rotación $\mathbf{F}(x, y) = (-y,\, x)$.

```python
# Malla de puntos
x_vals = np.linspace(-3, 3, 15)
y_vals = np.linspace(-3, 3, 15)
X, Y = np.meshgrid(x_vals, y_vals)

# Componentes del campo vectorial F(x, y) = (-y, x)
U = -Y    # componente horizontal
V =  X    # componente vertical

# Normalizar para visualizar solo la dirección
norma = np.sqrt(U**2 + V**2)
norma[norma == 0] = 1         # evitar división por cero
U_n, V_n = U / norma, V / norma

# Graficar
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

axes[0].quiver(X, Y, U, V, color='#1957d1', alpha=0.8)
axes[0].set_title(r'$\mathbf{F}(x,y) = (-y,\; x)$', fontsize=13)
axes[0].set_xlabel(r'$x$'); axes[0].set_ylabel(r'$y$')
axes[0].set_aspect('equal'); axes[0].grid(True, alpha=0.3)

axes[1].quiver(X, Y, U_n, V_n, color='#1957d1', alpha=0.8)
axes[1].set_title(r'Campo normalizado', fontsize=13)
axes[1].set_xlabel(r'$x$'); axes[1].set_ylabel(r'$y$')
axes[1].set_aspect('equal'); axes[1].grid(True, alpha=0.3)

plt.suptitle('Campo de rotación', fontsize=14)
plt.tight_layout()
plt.show()
```
